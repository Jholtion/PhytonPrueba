```python
import  pandas as pd

df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
df.head()
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip().str.title().str.replace(" ", "")
df.info()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
df["BaseMsrp"].fillna(df["BaseMsrp"].mean(), inplace=True)
df.dropna(subset=["ElectricRange"], inplace=True)
df.isnull().sum()
```

```python
df.dropna(subset=["LegislativeDistrict"], inplace=True)
df.dropna(subset=["VehicleLocation"], inplace=True)
df.dropna(subset=["ElectricUtility"], inplace=True)
df.dropna(subset=["2020CensusTract"], inplace=True)

df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
CantidadPorAo = df["ModelYear"].value_counts()
sns.barplot(x=CantidadPorAo.index, y=CantidadPorAo.values)
plt.xlabel("AO")
plt.
plt.ylabel("CANTIDAD DE VEHICULOS")
plt.title("VEHICULOS POR AO")
plt.show()
```
